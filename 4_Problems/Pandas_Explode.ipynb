```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

# Bar Plot of Top Skills for Cloud Engineers

```python
cloud_eng = df[df['job_title_short'] == 'Cloud Engineer']
```

```python
df_cloud_exploded = cloud_eng.explode('job_skills')

df_cloud_exploded['job_skills'].value_counts()
```

```python
df_cloud_exploded['job_skills'].value_counts().head(10).plot(kind='bar')

plt.title('Top Skills for Cloud Engineers')
plt.xlabel('Count of skills')
plt.ylabel('')


plt.show()
```
